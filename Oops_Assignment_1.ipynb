Problem 1: Bank Account create a class representing a bank account with attributes like account number, account holder name, and balance. Impelement methods to deposit and withdraw money from the account.

    Solution:

```python
import logging as log

log.basicConfig(
    filename = "BankAccount.log",
    level = log.DEBUG,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class BankAccount:
    """
    A simple calss for user account
    """
    
    def __init__(self, acct_no, acct_holder_name, balance):
        """
        Initialises attributes for opening user's profile
        """    
        self.acct_number = acct_no
        self.acct_holder_name = acct_holder_name
        self.balance = balance
        log.info(f"Inside Profile page- Acc No. : {self.acct_number}, Acc Holder's Name: {self.acct_holder_name}, Available Balance: ${self.balance}")
    
    
    def deposit(self, amount):
        """
        Deposit money in a user's account
        """
        log.info(f"Entered amount for deposit: ${amount}")
        try:
            if amount <= 0:
                raise ValueError("Invalid deposit amount. Amount must be positive.")
            self.balance += amount
        except ValueError as v:
            log.error(v)
        except Exception as e:
            log.error(f"An unexpected error has occured: {str(e)}")
        else:
            log.info(f"Deposit of {amount} is successful. New balance: ${self.balance}")        
         
    
    def withdraw(self, amount):
        """
        Function for withdrawing money from a user's account
        """
        log.info(f"Entered amount for withdraw: ${amount}")
        try:
            if amount > self.balance:
                raise ValueError("Transaction declined due to insufficient Balance.")
            elif amount <= 0:
                raise ValueError("Invalid withdraw amount, Amount must be positive")
            self.balance -= amount
        except ValueError as v:
            log.error(v)
        except Exception as e:
            log.error(f"An unexpected errro has occured: {str(e)}")
        else:
            log.info(f"Withdrawn of ${amount} is successful. New balance: ${self.balance}")
```

```python
# Example usage
user = BankAccount(12345, "Sahil Kumar Mandal", 500)
user.deposit(500)
user.withdraw(200)
```

Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implementing to rent a vehicle, return a vehicle, and display available vehicles.

    Solution:

```python
import logging as log

log.basicConfig(
    filename = "VehicleRentalSystem.log",
    level = log.INFO,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class VehicleRentalSystem:

    def __init__(self):
        """
        Inititalize the vehicle rental system with available and rented vehicles.
        """
        self.available_vehicles = ["Lamborghini", "Toyota Camry", "Jaguar", "BMW", "Fortuner"]
        self.rented_vehicles = []
        log.info(f"We have available vehicles: {self.available_vehicles} and Rented Vehicles: {self.rented_vehicles}")
    
    
    def add_vehicle(self, vehicles):
        """
        Add a list of vehicles to the available vehicles.
        """
        if isinstance(vehicles, list):
            log.info(f"Adding new vehicles: {vehicles}")
            self.available_vehicles.extend(vehicles)
            log.info(f"Now, Available vehicles: {self.available_vehicles}")
        else:
            log.error("Please provide a list of vehicles to add")
    
    
    def rent_vehicle(self, vehicle):
        """
        Rent a vehicle.
        """
        if vehicle in self.available_vehicles:
            log.info("Renting vehicle: %s", vehicle)
            self.rented_vehicles.append(vehicle)
            self.available_vehicles.remove(vehicle)
            log.info(f"Vehicle '{vehicle}' rented succesfully")
        elif vehicle in self.rented_vehicles:
            log.error("Vehicle is already rented, please check for another vehicle")
        else:
            log.error("Vehicle is not available, please check for another vehicle")
    
    
    def return_vehicle(self, vehicle):
        """
        Return a rented vehicle.
        """
        if vehicle in self.rented_vehicles:
            log.info("Returning vehicle: %s", vehicle)
            self.rented_vehicles.remove(vehicle)
            self.available_vehicles.append(vehicle)
            log.info(f"Vehicle '{vehicle}' returned succesfully")
        elif vehicle not in self.rented_vehicles and self.available_vehicles:
            log.warning("Vehicle doesn't belong to us or is not rented")
        

    def display_available_vehicles(self):
        """
        Display the list of available vehicles.
        """
        log.info("Checking available vehicles list")
        log.info(f"Available vehicles: {self.available_vehicles}")
```

```python
# Example usage
customer = VehicleRentalSystem()
customer.display_available_vehicles()
customer.rent_vehicle('Lamborghini')
customer.return_vehicle("Lamborghini")
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

    Solution:

```python
import logging as log

log.basicConfig(
    filename = "LibraryCatalog.log",
    level = log.DEBUG,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class LibraryCatalog:

    def __init__(self):
        self.available_books = ["Data science", "Algebra", "Calculas"]
        self.borrowed_books = []
        log.info(f"Available books: {self.available_books} and Borrowed books: {self.borrowed_books}")

    
    def add_books(self, new_books):
        """
        Add new books to the library catalog.
        """
        if isinstance(new_books, list):
            log.info(f"Adding new books in library: {new_books}")
            self.available_books.extend(new_books)
    
    
    def borrow_book(self, book):
        """
        Borrow a book from the library.
        """
        log.info("Borow books: %s", book)
        log.info("Checking availability")
        if book in self.available_books:
            log.info("Book is available")
            self.borrowed_books.append(book)
            self.available_books.remove(book)
            log.info("Removing book from the available list and added to the borrowed list")
        elif book in self.borrowed_books:
            log.error("This book is already borrowed")
        else:
            log.error("Book is not available")
        
    
    def display_available_books(self):
        """
        Display the list of available books in the library.
        """
        log.info("Available books: %s", self.available_books)
```

```python
# Example usage
catalog = LibraryCatalog()
catalog.add_books(["Python Programming", "Machine Learning"])
catalog.borrow_book("Machine Learning")
catalog.display_available_books()
```

Problem 5: Product Inventory create classes representing a product and an inventory system. Implement method to add products to the inventory, update product quantity, and display available products.

    Solution:

```python
import logging

logging.basicConfig(
    filename = "ProductInventory.log",
    level = log.DEBUG,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class Product:
    def __init__(self, product_id, product_name, product_price, product_quantity):
        self.product_id = product_id
        self.product_name = product_name
        self.product_price = product_price
        self.product_quantity = product_quantity
        logging.info(f"Product to add - Product ID: {self.product_id}, Product Name: {self.product_name}, Product Price: ${self.product_price}, Product Quantity: {self.product_quantity}")

    def __str__(self):
        return f"Product ID: {self.product_id}, Product Name: {self.product_name}, Product Price: ${self.product_price}, Product Quantity: {self.product_quantity}"
        
class Inventory:
   
    def __init__(self):
        self.products = []
        logging.info("Available products: %s", self.products)
    
    def add_product(self, product):
        """
        Adding product to the inventory
        """
        logging.info("Adding it to inventory: %s", product)
        self.products.append(product)

    def update_quantity(self, product_id, new_quantity):
        """
        Update the quantity of products in the inventory
        """
        logging.info(f"Updating product id no.{product_id} quantity, new quantity is {new_quantity}")
        for product in self.products:
            if product.product_id == product_id:
                product.product_quantity = new_quantity
                logging.info("Quantity updated successfully")
                break
            else:
                logging.error("Product not found")
    
    def display_products(self):
        """
        Display the list of available products in inventory
        """
        logging.info("Checking inventory products status:")
        if not self.products:
            logging.warning("Inventory is empty")
        else:
            logging.info("Current status: %s", [product.__dict__ for product in self.products])
```

```python
# Example usage

# Adding product to the inventory
product1 = Product(1, "Smart watch", 4000, 10)
product2 = Product(2, 'Pc', 1000000, 20)
product3 = Product(3, 'Shoes', 300, 40)

inventory = Inventory()
inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

# Displaying available products in inventory
inventory.display_products()

# Updating product quantity
inventory.update_quantity(2, 400)

# Displaying available products after updating quantity
inventory.display_products
```

Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and student details.

    Solution:

```python
import logging

logging.basicConfig(
    filename = "Student.log",
    level = logging.DEBUG,
    format = "%(asctime)s - %(levelname)s - %(message)s" 
)

class Student:

    def __init__(self, student_id, name, grades):
        """
        Initialize student details
        """
        try:
            self.student_id = student_id
            self.name = name
            self.grades = grades
            logging.info(f"User has entered student details: Student ID: {self.student_id}, Name: {self.name}, Grades: {self.grades}")
        except Exception as e:
            logging.error(f"Error while inititalizing student details: {e}")

    def average_grade(self):
        """
        Calculate average grade
        """
        try:
            logging.info("User is calculating average grade:")
            avg_grade = sum(self.grades) / len(self.grades)
            logging.info("Average grade: %f", avg_grade)
            return avg_grade
        except ZeroDivisionError:
            logging.error("ZeroDivisionError: cannot calculate average grade with zero grades")
            return 0
        
    def student_details(self):
        """
        Display the student details
        """
        try:
            logging.info("User want to print student details")
            logging.info(f"STUDENT DETAILS: Student ID: {self.student_id}, Name: {self.name}, Grades: {self.grades}, Average grade: {self.average_grade()}")
        except Exception as e:
            logging.error(f"Error while displaying student details: {e}")
```

```python
# Example usage
student = Student(123456, 'Sahil Kumar Mandal', [23, 45, 67, 89, 90])
student.average_grade()
student.student_details()
```

Problem 9: Social Media Profile Create a class representing a social media profiles with attributes like username and posts. Implement methods to add posts, display posts, and serch for posts by keyword.

    Solution:

```python
import logging

logging.basicConfig(
    filename = 'SocialMediaProfile.log',
    level = logging.DEBUG,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class SocialMediaProfile:

    def __init__(self,username):
        """
        Opening the social media profile from username
        """
        self.username = username
        self.posts = []
        logging.info(f"Inside social media profile: Username: {self.username}")

    def add_post(self, post_text):
        """
        Uploading post on the social media profile.
        """
        self.posts.append(post_text)
        logging.info("Uploading post: %s", post_text)

    def search_posts(self, keyword):
        """
        Searching post by keyword
        """
        logging.info("Searching post containing the keyword: %s", keyword)
        matching_posts = [post for post in self.posts if keyword in post]
        if not matching_posts:
            logging.error("No posts found containing the keyword: %s", keyword)
        else:
            logging.info(f"Posts containing the keyword '{keyword}':")
            for i, post in enumerate(matching_posts, start = 1):
                logging.info(f"{i}. {post}")

    def display_posts(self):
        """
        Displaying uploaded post
        """
        logging.info(f"Displaying posts: {self.posts}")
```

```python
# Example Usage
my_profile = SocialMediaProfile("Sahil Kumar Mandal")
my_profile.add_post("Today is having a great day")
my_profile.add_post("I'm having fun on the beack")
my_profile.add_post("Lets's celebrate Mr. Ramanujan birthday")
my_profile.display_posts()
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

    Solution:

```python
import logging as log

log.basicConfig(
    filename = 'TodoList.log',
    level = log.DEBUG,
    format = '%(asctime)s - %(levelname)s - %(message)s'
)

class TodoList:
    def __init__(self, task, due_date):
        '''
        Initialize the TodoList with a task and its due date.
        '''
        self.tasks = [{'Task': task, 'Due date': due_date, 'Completed': False}]
        log.info(f"Todolist created with task: {task}, due_date: {due_date}")
    
    
    def add_task(self, task, due_date):
        """
        Add a new task to the Todolist.
        """
        self.tasks.append({'Task': task, 'Due date': due_date, 'Completed': False})
        log.info(f"Task '{task}' added with due date: {due_date}")


    
    def mark_as_completed(self, task_index):
        """
        Mark a task as completed by its index.
        """   
        try:
            if 1 <= task_index <= len(self.tasks):
                task = self.tasks[task_index - 1]
                task['Completed'] = True
                log.info(f"Task '{task['Task']}' marked as completed successfully")
            else:
                log.error("Invalid task index: Index out of range")

        except Exception as e:
            log.error(f"An error occured: {str(e)}")
        
    
    def display_pending_tasks(self):
        """
        Displaying pending tasks.
        """
        pending_task = [task for task in self.tasks if not task["Completed"]]
        if pending_task:
            for i, task in enumerate(pending_task, start = 1):
                log.info(f"{i}. {task['Task']} (Due: {task['Due date']})")
        else:
            log.info("No pending tasks")
```

```python
my_todo_list = TodoList("Data structure and algorithm assignment", "2023-12-12")
my_todo_list.add_task("OOPs programming practice", "2023-12-11")
my_todo_list.add_task('Trip to Haridwar', '2023-11-23')
my_todo_list.display_pending_tasks()
my_todo_list.mark_as_completed(3)
```
